```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1080 = pd.read_csv("apristin-youtube[1080p60]-1x-novpn-windows-noisy-20201026.csv")
df720 = pd.read_csv("apristin-youtube[720p60]-1x-novpn-windows-noisy-20201026.csv")
df240 = pd.read_csv("apristin-youtube[240p]-1x-novpn-windows-noisy-20201026.csv")
df144 = pd.read_csv("apristin-youtube[144p]-1x-novpn-windows-noisy-20201026.csv")
dfNet = pd.read_csv("apristin-netflix-1x-novpn-windows-noisy-20201026.csv")
```

```python
vals = [df144, df240, df720, df1080]
titles = ["144p", "240p", "720p60", "1080p60"]
fig, ax = plt.subplots(figsize=(20, 10))
for k in range(4):
    plt.subplot(2, 2, k+1)
    plt.ylabel("Bytes")
    plt.title(titles[k])
    plt.plot(vals[k]["1->2Bytes"])
```
